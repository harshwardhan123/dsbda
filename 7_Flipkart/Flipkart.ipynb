```python
#import libraries------>
#1. requests - set a HTTp request to specified URL.It uses to request access.
#2. bs4 - Pulling data out of/from HTML and XML files/Websites.
import requests
import bs4
```

```python
#Geting Data request--------->
Requests1=requests.get('https://www.flipkart.com/flipkart-smartbuy-rs-g38c-bl-rs-acoustic-guitar-linden-wood/p/itm2b193bc5e7bd4?pid=ACGG4ABYHFSEKXJH&lid=LSTACGG4ABYHFSEKXJHHYQOPV&marketplace=FLIPKART&q=guitar&store=ypu%2Fujd&srno=s_1_10&otracker=search&otracker1=search&fm=Search&iid=cef486b6-4a89-4c32-bd82-b51ed9416a3f.ACGG4ABYHFSEKXJH.SEARCH&ppt=sp&ppn=sp&ssid=t4mgdx6gk00000001685543093724&qH=4c39e90d6a5c38a3')
Requests1
```

```python
#to see the content --------->
Requests1.content
```

```python
#BeautifulSoup - is a Method inside the bs4 library.
Guitar=bs4.BeautifulSoup(Requests1.text)
Guitar
```

```python
#Fetching Reviews/Comments-------->
Reviews = Guitar.find_all('div',{'class':'t-ZTKy'});
for review in Reviews:
    print(review.get_text()+"\n\n")
```

```python
#Fetching Rating-------->
#Average Overall Rating -
Ratings=Guitar.find('div',{'class':'_2d4LTz'}).get_text();
print(Ratings)
```

```python
#Individual Rating-------->
Individual_ratings=Guitar.find_all('div',{'class':'_3LWZlK _1BLPMq'});
for indi_ratings in Individual_ratings:
    print(indi_ratings.get_text()+'\n')
```

```python
#Fetching Tags-------->
Tags=Guitar.find('span',{'class':'B_NuCI'}).get_text();
print(Tags)
```

```python
#Fetching Customer Names--------->
Customer_name=Guitar.find_all('p',{'class':'_2sc7ZR _2V5EHH'});
for cust_name in Customer_name:
    print(cust_name.get_text()+"\n")
```

```python
#Exta Information--------->
#Fetching Price of Product-
Price=Guitar.find('div',{'class':'_30jeq3 _16Jk6d'}).get_text()
print(Price)
```

```python
#Fetching Specifications of Product--------->
Specifications=Guitar.find_all('div',{'class':'_3dtsli'});
for specifications in Specifications:
    print(specifications.get_text()+'\n')
```

```python
#Fetching Questions and Answers--------->
Questions=Guitar.find_all('div',{'class':'_1RWRBu'});
for questions in Questions:
    print(questions.get_text()+'\n')
```
